```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img_color = cv2.imread("cat.png", cv2.IMREAD_COLOR)

# OpenCV đọc ảnh theo BGR → đổi sang RGB để hiển thị đúng màu
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img_color)
plt.title("Ảnh gốc (màu)")
plt.axis("off")
plt.show()
```

```python
img_gray = cv2.cvtColor(img_color, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(4,4))
plt.imshow(img_gray, cmap='gray')
plt.title("Ảnh xám")
plt.axis("off")
plt.show()
```

```python
k = 50

bright_img = img_color.astype(np.int16) + k
bright_img = np.clip(bright_img, 0, 255).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(bright_img)
plt.title("Ảnh tăng độ sáng (+50)")
plt.axis("off")
plt.show()
```

```python
k = 50

dark_img = img_color.astype(np.int16) - k
dark_img = np.clip(dark_img, 0, 255).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(dark_img)
plt.title("Ảnh giảm độ sáng (-50)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_color)
plt.title("Gốc")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bright_img)
plt.title("Tăng sáng")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(dark_img)
plt.title("Giảm sáng")
plt.axis("off")

plt.show()
```
