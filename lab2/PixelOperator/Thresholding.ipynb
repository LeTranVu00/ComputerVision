```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread("cat.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
# Tnh nh cng  (intensity)
intensity = (
    0.299 * img[:,:,0] +
    0.587 * img[:,:,1] +
    0.114 * img[:,:,2]
).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(intensity, cmap='gray')
plt.title("nh cng  (Intensity)")
plt.axis("off")
plt.show()
```

```python
T = 128  # ngng

binary_img = np.where(intensity >= T, 255, 0).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(binary_img, cmap='gray')
plt.title("nh nh phn (Threshold = 128)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("nh mu gc")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(intensity, cmap='gray')
plt.title("Cng  sng")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(binary_img, cmap='gray')
plt.title("nh nh phn")
plt.axis("off")

plt.show()
```
