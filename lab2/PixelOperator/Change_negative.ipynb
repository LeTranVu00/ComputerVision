```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread("cat.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
# To nh m bn (negative) bng cch ly 255 tr i gi tr pixel ban u ca nh gc
negative_img = 255 - img

plt.figure(figsize=(4,4))
plt.imshow(negative_img)
plt.title("nh m bn (Negative)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Gc")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(negative_img)
plt.title("m bn")
plt.axis("off")

plt.show()
```
