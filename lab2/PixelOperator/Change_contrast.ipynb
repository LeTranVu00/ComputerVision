```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread("cat.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
alpha = 1.5  # h s tng phn (>1 l tng)

contrast_up = img.astype(np.float32) * alpha
contrast_up = np.clip(contrast_up, 0, 255).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(contrast_up)
plt.title("nh tng tng phn ( = 1.5)")
plt.axis("off")
plt.show()
```

```python
alpha = 0.6  # h s tng phn (<1 l gim)

contrast_down = img.astype(np.float32) * alpha
contrast_down = np.clip(contrast_down, 0, 255).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(contrast_down)
plt.title("nh gim tng phn ( = 0.6)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Gc")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(contrast_up)
plt.title("Tng tng phn")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(contrast_down)
plt.title("Gim tng phn")
plt.axis("off")

plt.show()
```
