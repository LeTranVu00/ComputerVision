```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("cat.png")
if img is None:
    raise FileNotFoundError("Khng tm thy nh cat.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
mean_img = cv2.blur(img, (5,5))

plt.imshow(mean_img)
plt.title("Lc trung bnh (Mean)")
plt.axis("off")
plt.show()
```

```python
gaussian_img = cv2.GaussianBlur(img, (5,5), 0)

plt.imshow(gaussian_img)
plt.title("Lc Gaussian")
plt.axis("off")
plt.show()
```

```python
median_img = cv2.medianBlur(img, 5)

plt.imshow(median_img)
plt.title("Lc Median")
plt.axis("off")
plt.show()
```

```python
kernel_sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

sharpen_img = cv2.filter2D(img, -1, kernel_sharpen)

plt.imshow(sharpen_img)
plt.title("Lm sc nt")
plt.axis("off")
plt.show()
```

```python
kernel_edge = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

edge_img = cv2.filter2D(img, -1, kernel_edge)

plt.imshow(edge_img)
plt.title("Pht hin cnh")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(14,8))

titles = [
    "Gc",
    "Mean",
    "Gaussian",
    "Median",
    "Sharpen",
    "Edge"
]

images = [
    img,
    mean_img,
    gaussian_img,
    median_img,
    sharpen_img,
    edge_img
]

for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
