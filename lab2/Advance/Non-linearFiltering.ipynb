```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("cat.png")
if img is None:
    raise FileNotFoundError("Khng tm thy nh cat.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
noisy = img.copy()
h, w, _ = noisy.shape

num_noise = int(0.01 * h * w)
for _ in range(num_noise):
    x = np.random.randint(0, h)
    y = np.random.randint(0, w)
    noisy[x, y] = [0, 0, 0] if np.random.rand() < 0.5 else [255, 255, 255]

plt.imshow(noisy)
plt.title("nh nhiu mui tiu")
plt.axis("off")
plt.show()
```

```python
median_img = cv2.medianBlur(noisy, 5)

plt.imshow(median_img)
plt.title("Lc Median")
plt.axis("off")
plt.show()
```

```python
bilateral_img = cv2.bilateralFilter(
    img,
    d=9,
    sigmaColor=75,
    sigmaSpace=75
)

plt.imshow(bilateral_img)
plt.title("Lc Bilateral")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(noisy)
plt.title("nh nhiu")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(median_img)
plt.title("Median")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(bilateral_img)
plt.title("Bilateral")
plt.axis("off")

plt.show()
```
