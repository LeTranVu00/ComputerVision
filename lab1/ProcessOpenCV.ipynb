```python
import cv2
import matplotlib.pyplot as plt

# c nh
mrgold = cv2.imread('mr_gold.png')

# Hm h tr hin th nhanh bng plt
def show_image(img, title="Image", is_gray=False):
    plt.figure(figsize=(6, 4))
    if is_gray:
        plt.imshow(img, cmap='gray')
    else:
        # OpenCV dng BGR, Matplotlib dng RGB nn cn chuyn i
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
    plt.title(title)
    plt.axis('off') # n trc ta 
    plt.show()

# Hin th nh gc
show_image(mrgold, 'Mr Gold')
```

```python
# chuyn i khng gian mu sang GRAY
gray_mrgold = cv2.cvtColor(mrgold, cv2.COLOR_BGR2GRAY)
show_image(gray_mrgold, 'Gray Mr Gold', is_gray=True)
```

```python
# chuyn i sang khng gian mu HSV
hsv_mrgold = cv2.cvtColor(mrgold, cv2.COLOR_BGR2HSV)
show_image(hsv_mrgold, 'HSV Mr Gold')
```

```python
# chuyn i sang khng gian mu LAB
lab_mrgold = cv2.cvtColor(mrgold, cv2.COLOR_BGR2LAB)
show_image(lab_mrgold, 'LAB Mr Gold')
```

```python
# ct xn nh
anh_cat = mrgold[100:400, 200:500]
show_image(anh_cat, 'Cropped Mr Gold')
```

```python
# thay i kch thc nh
resized_mrgold = cv2.resize(mrgold, (300, 400))
show_image(resized_mrgold, 'Resized Mr Gold')
```

```python
# thay i kch thc nh (cch 2)
width = 200
height = 300
dim = (width, height)
fixed_size_mrgold = cv2.resize(mrgold, dim)
show_image(fixed_size_mrgold, 'Fixed Size Mr Gold')
```

```python
# v hnh v ch ln nh
anh_ve = mrgold.copy()

# v ng thng
cv2.line(anh_ve, (50, 50), (200, 50), (0, 0, 255), 5)

# v hnh trn
cv2.circle(anh_ve, (100, 150), 40, (0, 0, 255), 10)

# v hnh ch nht
cv2.rectangle(anh_ve, (200, 200), (300, 300), (255, 0, 0), 3)

# v ch
cv2.putText(anh_ve, 'Cau Vang', (50, 50),
            cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

show_image(anh_ve, 'Drawn Mr Gold')
```

```python
# lu nh ra file
cv2.imwrite('mr_gold.jpg', mrgold)
```
