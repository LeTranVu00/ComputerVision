```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("cat.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
kernel_sharpen = np.array([
    [0, -1,  0],
    [-1, 5, -1],
    [0, -1,  0]
])

sharpen_img = cv2.filter2D(img, -1, kernel_sharpen)

plt.figure(figsize=(4,4))
plt.imshow(sharpen_img)
plt.title("nh lm sc nt")
plt.axis("off")
plt.show()
```

```python
kernel_strong = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])

sharpen_strong = cv2.filter2D(img, -1, kernel_strong)

plt.figure(figsize=(4,4))
plt.imshow(sharpen_strong)
plt.title("nh sc nt mnh")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Gc")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(sharpen_img)
plt.title("Sharpen thng")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sharpen_strong)
plt.title("Sharpen mnh")
plt.axis("off")

plt.show()
```
