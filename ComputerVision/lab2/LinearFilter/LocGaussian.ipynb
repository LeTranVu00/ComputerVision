```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("cat.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
gaussian_3x3 = cv2.GaussianBlur(img, (3, 3), sigmaX=0)

plt.figure(figsize=(4,4))
plt.imshow(gaussian_3x3)
plt.title("Gaussian 33")
plt.axis("off")
plt.show()
```

```python
gaussian_5x5 = cv2.GaussianBlur(img, (5, 5), sigmaX=0)

plt.figure(figsize=(4,4))
plt.imshow(gaussian_5x5)
plt.title("Gaussian 55")
plt.axis("off")
plt.show()
```

```python
mean_5x5 = cv2.blur(img, (5, 5))

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Gc")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(mean_5x5)
plt.title("Mean 55")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(gaussian_5x5)
plt.title("Gaussian 55")
plt.axis("off")

plt.show()
```
