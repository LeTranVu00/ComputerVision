```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("cat.png")
if img is None:
    raise FileNotFoundError("Khng tm thy nh cat.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("nh gc")
plt.axis("off")
plt.show()
```

```python
kernel_blur = np.ones((3,3), dtype=np.float32) / 9
# -1 l  gi chiu su nh gc (kiu d liu)
blur_custom = cv2.filter2D(img, -1, kernel_blur)

plt.imshow(blur_custom)
plt.title("Kernel t thit k  Lm m")
plt.axis("off")
plt.show()
```

```python
kernel_sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

sharpen_custom = cv2.filter2D(img, -1, kernel_sharpen)

plt.imshow(sharpen_custom)
plt.title("Kernel t thit k  Sc nt")
plt.axis("off")
plt.show()
```

```python
kernel_edge = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

edge_custom = cv2.filter2D(img, -1, kernel_edge)

plt.imshow(edge_custom)
plt.title("Kernel t thit k  Pht hin cnh")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12,8))

titles = ["Gc", "Blur", "Sharpen", "Edge"]
images = [img, blur_custom, sharpen_custom, edge_custom]

for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
