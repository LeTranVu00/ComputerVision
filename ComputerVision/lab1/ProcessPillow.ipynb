```python
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

# --- 0. Khi to & c nh ---
CauVang = 'mr_gold.png'

mrgold = Image.open(CauVang).convert("RGB")
anh_goc = mrgold.copy()

#  hin th ngay bn di, ch cn gi tn bin hoc dng display()
display(mrgold)
```

```python
# --- 1. Chuyn i khng gian mu ---
gray_mrgold = anh_goc.convert('L')
display(gray_mrgold)
```

```python
# --- 2. Ct xn nh ---
# Pillow: (left=200, top=100, right=500, bottom=400)
anh_cat = anh_goc.crop((200, 100, 500, 400))
display(anh_cat)
```

```python
# --- 3 & 4. Thay i kch thc nh ---
resized_mrgold_1 = anh_goc.resize((300, 400))
display(resized_mrgold_1)
```

```python
# --- 4. Thay i kch thc nh (200x300) ---
width = 200
height = 300
fixed_size_mrgold = anh_goc.resize((width, height))
display(fixed_size_mrgold)
```

```python
# --- 5. V hnh v ch ln nh ---
anh_ve = anh_goc.copy()
draw = ImageDraw.Draw(anh_ve)

mau_do = (255, 0, 0)
mau_xanh_duong = (0, 0, 255)

# V cc hnh khi
draw.line([(50, 50), (200, 50)], fill=mau_do, width=5)
draw.ellipse((60, 110, 140, 190), outline=mau_do, width=5)
draw.rectangle([(200, 200), (300, 300)], outline=mau_xanh_duong, width=3)

# V ch
try:
    font = ImageFont.truetype("arial.ttf", size=30)
except IOError:
    font = ImageFont.load_default()

draw.text((50, 50), 'Cau Vang', fill=mau_xanh_duong, font=font)

# Hin th kt qu cui cng
display(anh_ve)
```

```python
# --- 6. Lu nh ---
mrgold.save('mr_gold_pillow.jpg')
print(" lu nh thnh mr_gold_pillow.jpg")
```
